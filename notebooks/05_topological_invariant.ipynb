# Objective

The objective of this notebook is to compute a topological invariant that distinguishes between the topological and trivial phases of the Kitaev chain.

Unlike the previous notebooks, which focused on observable quantities such as energy spectra and localization, this notebook introduces a mathematical quantity that characterizes the global topology of the system.

The invariant allows us to identify the phase of the system without explicitly examining the eigenstates.

# Theoretical Background

For the Kitaev chain, the phase boundary occurs when

$$
|\mu| = 2t
$$

The topological invariant can be expressed as

$$
Q = \text{sign}
\Big[
(\mu + 2t)
(\mu - 2t)
\Big]
$$

The value of Q distinguishes the two phases:

- Q = -1 : Topological Phase
- Q = +1 : Trivial Phase

The invariant changes only when the bulk energy gap closes, making it a robust identifier of topological order.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 1.0

mu_values = np.linspace(
    -4,
    4,
    400
)

invariant = []

for mu in mu_values:

    Q = np.sign(
        (mu + 2*t)
        *
        (mu - 2*t)
    )

    invariant.append(Q)

invariant = np.array(invariant)
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    mu_values,
    invariant,
    linewidth=2
)

plt.axvline(
    x=-2,
    linestyle="--",
    label="μ = -2"
)

plt.axvline(
    x=2,
    linestyle="--",
    label="μ = +2"
)

plt.xlabel("Chemical Potential μ")
plt.ylabel("Topological Invariant Q")

plt.title(
    "Topological Invariant of the Kitaev Chain"
)

plt.grid(True)

plt.legend()

plt.show()
```

# Results

The topological invariant was evaluated across the interval

$$
-4 \le \mu \le 4
$$

allowing the phase structure of the Kitaev chain to be identified directly from a topological quantity.

# Observations

The topological invariant remains constant within each phase and changes abruptly at

$$
\mu = -2
$$

and

$$
\mu = +2
$$

which coincide with the phase boundaries observed in the energy-spectrum calculations.

Within the interval

$$
-2 < \mu < 2
$$

the invariant takes the value

$$
Q = -1
$$

indicating the topological superconducting phase.

Outside this interval, the invariant changes to

$$
Q = +1
$$

corresponding to the trivial phase.

The phase boundaries identified by the invariant match exactly with the locations where Majorana zero modes appear and disappear.

This agreement demonstrates the direct relationship between topological order and Majorana edge states in the Kitaev chain.

# Conclusion

The topological invariant successfully distinguishes between the topological and trivial phases of the Kitaev chain.

The invariant changes only at the critical points where the bulk energy gap closes, confirming the topological nature of the phase transition.

The results establish a direct connection between the mathematical topology of the system and the physical emergence of Majorana zero modes.
