# Objective

The objective of this notebook is to study the effect of chain length on the energy of Majorana zero modes.

In finite systems, Majorana modes located at opposite ends of the chain can overlap with one another.

This overlap causes the zero-energy degeneracy to split slightly.

By increasing the chain length, the overlap decreases and the Majorana modes become increasingly protected.

# Theoretical Background

Majorana zero modes exist at the boundaries of the topological phase.

For a finite chain, the two edge modes are not perfectly independent.

Their wavefunctions overlap and produce a small energy splitting.

The overlap decreases approximately exponentially with chain length:

$$
E_{\mathrm{split}}
\propto
e^{-N/\xi}
$$

where:

- $N$ is the chain length
- $\xi$ is the localization length

As the chain becomes longer, the splitting approaches zero and the Majorana modes become topologically protected.

```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append("../src")

from hamiltonian import build_kitaev_chain
```

```python
t = 1.0
delta = 1.0
mu = 0.5

chain_lengths = [
    10,
    20,
    30,
    40,
    50,
    75,
    100
]

min_energies = []

for N in chain_lengths:

    H = build_kitaev_chain(
        N=N,
        t=t,
        delta=delta,
        mu=mu
    )

    eigenvalues, _ = np.linalg.eigh(H)

    min_energy = np.min(
        np.abs(eigenvalues)
    )

    min_energies.append(
        min_energy
    )

    print(
        f"N={N}, "
        f"Minimum Energy={min_energy:.8e}"
    )
```

```python
plt.figure(figsize=(10,6))

plt.semilogy(
    chain_lengths,
    min_energies,
    "o-"
)

plt.xlabel("Chain Length N")

plt.ylabel(
    "Minimum Absolute Energy"
)

plt.title(
    "Finite-Size Scaling of Majorana Energy Splitting"
)

plt.grid(True)

plt.show()
```

# Results

The minimum absolute energy was computed for several chain lengths while keeping the system within the topological phase.

The resulting curve illustrates how the energy splitting evolves as the distance between the Majorana modes increases.

# Observations

The minimum energy decreases rapidly as the chain length increases.

For N = 10, the minimum energy is approximately

$$
1.8 \times 10^{-6}
$$

while for N = 20 it decreases to approximately

$$
1.7 \times 10^{-12}
$$

and by N = 30 reaches values on the order of

$$
10^{-16}
$$

The rapid reduction in energy splitting indicates that the overlap between Majorana modes located at opposite ends of the chain decreases strongly with increasing system size.

For chain lengths larger than approximately N = 30, the minimum energy reaches the numerical precision limit of double-precision arithmetic.

The small fluctuations observed beyond this point are therefore numerical artifacts rather than physical effects.

These results provide strong evidence that the observed zero-energy states correspond to Majorana edge modes and demonstrate how topological protection improves as the chain length increases.

# Conclusion

The finite-size scaling analysis demonstrates that the Majorana energy splitting decreases rapidly as the chain length increases.

The results confirm that finite energy splittings originate from overlap between edge-localized Majorana modes.

As the chain becomes longer, the overlap vanishes and the zero-energy degeneracy is restored.

This behavior is a key feature of topological protection and is one of the primary reasons Majorana modes are of interest for fault-tolerant quantum computation.
