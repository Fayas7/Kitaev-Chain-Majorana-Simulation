This notebook investigates how the energy spectrum evolves as the chemical potential μ is varied and identifies the phase transition between topological and trivial superconducting phases.

# Objective

The objective of this notebook is to study the evolution of the Kitaev chain energy spectrum as a function of the chemical potential μ.

By varying μ, we can:

- Identify topological and trivial phases
- Observe gap closing and reopening
- Locate critical points
- Verify theoretical predictions

This analysis provides direct numerical evidence of a topological phase transition.

# Theoretical Background

For the Kitaev chain, the phase boundary is determined by

$$
|\mu| = 2t
$$

For:

$$
|\mu| < 2t
$$

the system is in the topological superconducting phase.

For:

$$
|\mu| > 2t
$$

the system is in the trivial phase.

At the critical points:

$$
\mu = \pm 2t
$$

the bulk energy gap closes, signaling a topological phase transition.

```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append("../src")

from hamiltonian import build_kitaev_chain
```

```python
N = 20
t = 1.0
delta = 1.0

mu_values = np.linspace(-4, 4, 200)

all_energies = []

for mu in mu_values:

    H = build_kitaev_chain(
        N=N,
        t=t,
        delta=delta,
        mu=mu
    )

    eigenvalues, _ = np.linalg.eigh(H)

    all_energies.append(eigenvalues)

all_energies = np.array(all_energies)
```

```python
plt.figure(figsize=(10,6))

for band in range(all_energies.shape[1]):

    plt.plot(
        mu_values,
        all_energies[:, band],
        linewidth=0.8
    )

plt.axvline(
    x=-2,
    linestyle="--",
    label="μ = -2"
)

plt.axvline(
    x=2,
    linestyle="--",
    label="μ = +2"
)

plt.xlabel("Chemical Potential μ")
plt.ylabel("Energy")

plt.title("Topological Phase Transition in the Kitaev Chain")

plt.grid(True)

plt.legend()

plt.show()
```

# Results

The complete energy spectrum was calculated over the interval

$$
-4 \leq \mu \leq 4
$$

allowing the evolution of all energy bands to be visualized as a function of the chemical potential.

# Observations

The energy spectrum exhibits clear particle-hole symmetry, with positive and negative energy branches appearing as mirror images about E = 0.

A pair of zero-energy states is observed throughout the interval

$$
-2 < \mu < 2
$$

forming a nearly flat band at zero energy.

The zero-energy states disappear at

$$
\mu = -2
$$

and

$$
\mu = +2
$$

which coincide with the theoretical phase boundaries predicted by the Kitaev chain model.

Near these critical points, the bulk energy bands approach the zero-energy states, indicating a closing of the energy gap.

The persistence of zero-energy modes within the interval

$$
-2 < \mu < 2
$$

provides numerical evidence for the topological superconducting phase and the existence of Majorana edge states.

Outside this interval, the system transitions into the trivial phase where the zero-energy modes are absent.

# Conclusion

The numerical calculations successfully reproduce the topological phase transition predicted by the Kitaev chain model.

The critical values

$$
\mu = \pm 2t
$$

separate the topological and trivial phases and coincide with the disappearance of zero-energy states.

This result establishes the connection between Majorana modes and the underlying topological structure of the system.
