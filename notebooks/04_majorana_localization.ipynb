This notebook investigates the spatial localization of zero-energy states in the Kitaev chain and verifies whether these states correspond to Majorana edge modes.

# Objective

The objective of this notebook is to determine where the zero-energy states are located within the Kitaev chain.

The previous notebooks established the existence of zero-energy modes and identified the topological phase region.

The next step is to verify whether these states are localized at the boundaries of the chain, as predicted for Majorana zero modes.

# Theoretical Background

Majorana zero modes are localized boundary excitations that appear in the topological phase of the Kitaev chain.

Unlike ordinary bulk states, Majorana modes are concentrated near the ends of the chain.

Therefore, observing edge localization provides strong evidence that the zero-energy states identified in the spectrum correspond to Majorana modes rather than ordinary excitations.

```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append("../src")

from hamiltonian import build_kitaev_chain
```

```python
N = 20
t = 1.0
delta = 1.0
mu = 0.0

H = build_kitaev_chain(
    N=N,
    t=t,
    delta=delta,
    mu=mu
)

eigenvalues, eigenvectors = np.linalg.eigh(H)
```

```python
zero_mode_index = np.argmin(
    np.abs(eigenvalues)
)

print(
    "Closest energy:",
    eigenvalues[zero_mode_index]
)

zero_mode = eigenvectors[:, zero_mode_index]
```

```python
particle_part = zero_mode[:N]
hole_part = zero_mode[N:]

probability = (
    np.abs(particle_part)**2 +
    np.abs(hole_part)**2
)

probability = (
    probability /
    np.sum(probability)
)

probability
```

```python
sites = np.arange(
    1,
    N + 1
)

plt.figure(figsize=(10,5))

plt.plot(
    sites,
    probability,
    "o-"
)

plt.xlabel("Site Index")
plt.ylabel("Probability Density")

plt.title(
    "Majorana Zero Mode Localization"
)

plt.grid(True)

plt.show()
```

# Results

The probability density of the zero-energy eigenstate was computed by combining the particle and hole contributions at each lattice site.

The resulting distribution reveals the spatial structure of the Majorana mode.

# Observations

The probability density is concentrated almost entirely at the final lattice site of the chain.

All remaining sites exhibit negligible probability density compared to the edge site.

This localization behavior is a characteristic signature of a Majorana edge mode and confirms that the zero-energy state identified in the energy spectrum is not a bulk excitation.

The strong localization observed in this calculation is a consequence of the parameter choice

μ = 0

and

t = Δ

which corresponds to the special point of the Kitaev chain where Majorana modes become perfectly localized at the boundaries.

The result provides direct numerical evidence for the existence of Majorana zero modes in the topological phase.

# Conclusion

The spatial profile of the zero-energy state demonstrates strong edge localization.

Combined with the energy-spectrum and phase-transition results, this confirms that the Kitaev chain supports Majorana zero modes within the topological phase.

The localization analysis provides one of the strongest numerical signatures of topological superconductivity in the model.
