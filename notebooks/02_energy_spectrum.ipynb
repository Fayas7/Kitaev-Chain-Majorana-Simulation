This notebook diagonalizes the Bogoliubov-de Gennes Hamiltonian of the Kitaev chain and analyzes the resulting energy spectrum.

## Objective

The objective of this notebook is to compute the energy spectrum of the Kitaev chain and investigate the existence of zero-energy states.

The energy spectrum provides important information about:

- Particle-hole symmetry
- Energy gaps
- Topological phases
- Majorana zero modes

By diagonalizing the Hamiltonian, we obtain the allowed energy levels of the system.

## Theoretical Background

The energy spectrum is obtained by solving the eigenvalue equation

$$
H\psi = E\psi
$$

where:

- H is the Bogoliubov-de Gennes Hamiltonian
- ψ is an eigenstate
- E is the corresponding energy

For superconducting systems described by the BdG formalism, the spectrum exhibits particle-hole symmetry. Consequently, positive and negative energy levels appear in symmetric pairs.

```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append("../src")

from hamiltonian import build_kitaev_chain
```

```python
N = 20
t = 1.0
delta = 1.0
mu = 0.0

H = build_kitaev_chain(
    N=N,
    t=t,
    delta=delta,
    mu=mu
)

eigenvalues, eigenvectors = np.linalg.eigh(H)

print("Lowest absolute energy:")
print(np.min(np.abs(eigenvalues)))
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    range(len(eigenvalues)),
    eigenvalues,
    "o"
)

plt.xlabel("State Index")
plt.ylabel("Energy")
plt.title("Energy Spectrum of the Kitaev Chain")

plt.grid(True)

plt.show()
```

## Results

The Hamiltonian was diagonalized successfully and the full energy spectrum was obtained.

The spectrum exhibits symmetry around E = 0, which is characteristic of particle-hole symmetry in superconducting systems.

## Observations

The energy spectrum was calculated for a chain with N = 20 sites and parameters t = 1.0, Δ = 1.0, and μ = 0.0.

The spectrum exhibits particle-hole symmetry, with positive and negative energy states appearing in symmetric pairs.

Two states were found extremely close to zero energy, with the minimum absolute energy equal to approximately

$$
4.44 \times 10^{-16}
$$

which is effectively zero within numerical precision.

The large degeneracy observed at energies ±2 is a consequence of the highly symmetric parameter choice μ = 0 and t = Δ.

The existence of near-zero-energy states provides the first numerical evidence for Majorana zero modes in the Kitaev chain.

# Conclusion

The energy spectrum confirms the expected particle-hole symmetry of the Bogoliubov-de Gennes Hamiltonian.

Near-zero-energy states were observed, suggesting the presence of Majorana edge modes and motivating further investigation of the topological properties of the system.
