## Theoretical Background

The Kitaev chain is a one-dimensional model of a p-wave topological superconductor.

The model contains three physical processes:

1. Chemical potential (μ)
2. Nearest-neighbor hopping (t)
3. Superconducting pairing (Δ)

To describe superconductivity, the Hamiltonian is written in the Bogoliubov-de Gennes (BdG) formalism, where particle and hole degrees of freedom are treated simultaneously.

For a chain containing N lattice sites, the resulting Hamiltonian has dimensions 2N × 2N.

## Kitaev Chain Hamiltonian

$$
H =
-\mu \sum_i c_i^\dagger c_i
-
t \sum_i
(c_i^\dagger c_{i+1}+h.c.)
+
\Delta \sum_i
(c_i c_{i+1}+h.c.)
$$
